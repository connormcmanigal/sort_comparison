## Plotting and Descriptive Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt

tr_df = pd.read_csv('uniform_results.csv')
as_df = pd.read_csv('almost_sorted_results.csv')
```

### MergeSort Time Analysis

```python
merge_tr_1000 = tr_df[(tr_df['Algorithm'] == 'MergeSort') & (tr_df['Data Size'] == 1000)]
merge_tr_2000 = tr_df[(tr_df['Algorithm'] == 'MergeSort') & (tr_df['Data Size'] == 2000)]
merge_tr_4000 = tr_df[(tr_df['Algorithm'] == 'MergeSort') & (tr_df['Data Size'] == 4000)]
merge_tr_8000 = tr_df[(tr_df['Algorithm'] == 'MergeSort') & (tr_df['Data Size'] == 8000)]
merge_tr_16000 = tr_df[(tr_df['Algorithm'] == 'MergeSort') & (tr_df['Data Size'] == 16000)]

merge_as_1000 = as_df[(as_df['Algorithm'] == 'MergeSort') & (as_df['Data Size'] == 1000)]
merge_as_2000 = as_df[(as_df['Algorithm'] == 'MergeSort') & (as_df['Data Size'] == 2000)]
merge_as_4000 = as_df[(as_df['Algorithm'] == 'MergeSort') & (as_df['Data Size'] == 4000)]
merge_as_8000 = as_df[(as_df['Algorithm'] == 'MergeSort') & (as_df['Data Size'] == 8000)]
merge_as_16000 = as_df[(as_df['Algorithm'] == 'MergeSort') & (as_df['Data Size'] == 16000)]
```

#### MergeSort: Descriptive Statistics

```python
pd.set_option('display.width', 500)
print("MergeSort Statistics on True Random Data")
merge_sort_tr = tr_df[tr_df['Algorithm'] == 'MergeSort']
merge_tr_stats = merge_sort_tr.groupby('Data Size').describe()
merge_tr_stats.drop(columns='Theoretical Big-O', inplace=True)
merge_tr_stats.reset_index(inplace=True)
merge_tr_stats.columns = merge_tr_stats.columns.map(' '.join)
merge_tr_stats = pd.DataFrame(merge_tr_stats)
print(merge_tr_stats, '\n')

print("MergeSort Statistics on Almost Sorted Data")
#merge_sort_as = as_df[as_df['Algorithm'] == 'MergeSort']
#merge_as_stats = merge_sort_as.groupby('Data Size').describe()
#merge_as_stats.drop(columns='Theoretical Big-O', inplace=True)
#merge_as_stats.reset_index(inplace=True)
#merge_as_stats.columns = merge_as_stats.columns.map(' '.join)
#merge_as_stats = pd.DataFrame(merge_as_stats)
#print(merge_as_stats)
```

#### Plotting MergeSort: True Random vs. Almost Sorted

```python
fig, axs = plt.subplots(1, 5, figsize=(30, 10))

axs[0].bar(range(len(merge_tr_1000)), merge_tr_1000['Runtime'], color="orange")
axs[0].set_title('Data Size 1000')
axs[0].set_ylim(0, 0.04)
axs[0].grid(True)

axs[1].bar(range(len(merge_tr_2000)), merge_tr_2000['Runtime'], color="orange")
axs[1].set_title('Data Size 2000')
axs[1].set_ylim(0, 0.04)
axs[1].grid(True)

axs[2].bar(range(len(merge_tr_4000)), merge_tr_4000['Runtime'], color="orange")
axs[2].set_title('Data Size 4000')
axs[2].set_ylim(0, 0.04)
axs[2].grid(True)

axs[3].bar(range(len(merge_tr_8000)), merge_tr_8000['Runtime'], color="orange")
axs[3].set_title('Data Size 8000')
axs[3].set_ylim(0, 0.04)
axs[3].grid(True)

axs[4].bar(range(len(merge_tr_16000)), merge_tr_16000['Runtime'], color="orange")
axs[4].set_title('Data Size 16000')
axs[4].set_ylim(0, 0.04)
axs[4].grid(True)

fig.supxlabel("Trial #", fontsize=20)
fig.supylabel("Runtime(seconds)", fontsize=20)
fig.suptitle("MergeSort on True Random Data", fontsize=25)
fig.subplots_adjust(left=0.05, right=0.95) 

plt.show()

fig, axs = plt.subplots(1, 5, figsize=(30, 10))

axs[0].bar(range(len(merge_as_1000)), merge_as_1000['Runtime'], color="orange")
axs[0].set_title('Data Size 1000')
axs[0].set_ylim(0, 0.04)
axs[0].grid(True)

axs[1].bar(range(len(merge_as_2000)), merge_as_2000['Runtime'], color="orange")
axs[1].set_title('Data Size 2000')
axs[1].set_ylim(0, 0.04)
axs[1].grid(True)

axs[2].bar(range(len(merge_as_4000)), merge_as_4000['Runtime'], color="orange")
axs[2].set_title('Data Size 4000')
axs[2].set_ylim(0, 0.04)
axs[2].grid(True)

axs[3].bar(range(len(merge_as_8000)), merge_as_8000['Runtime'], color="orange")
axs[3].set_title('Data Size 8000')
axs[3].set_ylim(0, 0.04)
axs[3].grid(True)

axs[4].bar(range(len(merge_as_16000)), merge_as_16000['Runtime'], color="orange")
axs[4].set_title('Data Size 16000')
axs[4].set_ylim(0, 0.04)
axs[4].grid(True)

fig.supxlabel("Trial #", fontsize=20)
fig.supylabel("Runtime(seconds)", fontsize=20)
fig.suptitle("MergeSort on Almost-sorted Data", fontsize=25)
fig.subplots_adjust(left=0.05, right=0.95) 

plt.show()
```

```python
#Size vs. Execution Time in One Plot
merge_tr_runtime = []
merge_as_runtime = []

working_mean = 0

merge_tr_runtime.append(merge_tr_1000['Runtime'].mean())
merge_tr_runtime.append(merge_tr_2000['Runtime'].mean())
merge_tr_runtime.append(merge_tr_4000['Runtime'].mean())
merge_tr_runtime.append(merge_tr_8000['Runtime'].mean())
merge_tr_runtime.append(merge_tr_16000['Runtime'].mean())

merge_as_runtime.append(merge_as_1000['Runtime'].mean())
merge_as_runtime.append(merge_as_2000['Runtime'].mean())
merge_as_runtime.append(merge_as_4000['Runtime'].mean())
merge_as_runtime.append(merge_as_8000['Runtime'].mean())
merge_as_runtime.append(merge_as_16000['Runtime'].mean())

# These are your n values
n_values = [1000, 2000, 4000, 8000, 16000]

plt.figure(figsize=(10, 5))  # Optional: Adjusts the figure size

# Plotting both lists
plt.plot(n_values, merge_tr_runtime, label='Merge TR runtime', marker='o')  # 'o' adds markers to each point
plt.plot(n_values, merge_as_runtime, label='Merge AS runtime', marker='o')

# Adding titles and labels
plt.title('Mergesort Runtimes, both True Random and Almost-sorted')
plt.xlabel('n')
plt.ylabel('Runtime')
plt.xticks(n_values, labels=[f'n = {val}' for val in n_values])  # Sets custom labels for the x-axis

# Adding a legend to explain which line is which
plt.legend()

# Display the plot
plt.show()
```
